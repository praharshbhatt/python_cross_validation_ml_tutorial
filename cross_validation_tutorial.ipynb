### Split the data into training and testing

```python
# 70% training and 30% test
from sklearn.model_selection import train_test_split

X = df.drop('diagnosis', axis=1)  # Features
y = df.diagnosis  # Target variables

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=1
)


print(X.shape)
print(y.shape)
```

```python
import numpy as np
from sklearn import model_selection
from sklearn import datasets
from sklearn import svm
import sklearn.ensemble as rf
import sklearn.neighbors as knn
import sklearn.tree as dt
import statistics

models = [
    ('RF', rf.RandomForestClassifier()),
    ('KNN', knn.KNeighborsClassifier()),
    ('DT', dt.DecisionTreeClassifier()),
]

result = {}
for name, model in models:
    k_fold = model_selection.KFold(n_splits=6, shuffle=True)
    cv_results = model_selection.cross_validate(
        model,
        X_train,
        y_train,
        cv=k_fold,
        scoring=['accuracy']
    )
    result[name] = statistics.mean(cv_results['test_accuracy'])
result
```

### Cross validation results and model creation
